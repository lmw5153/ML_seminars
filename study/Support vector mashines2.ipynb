# 1.  Lagrange duality

`-` Lagrange multipliers 

- $f(w)$를 minimize할 것인데, 조건이 있다.
- $h(w) = 0 $이 되면서 minimize를 할것(f를 미분했을 때 0이 아닌 지점이 됨)
- 이것은 우리가 라그랑주 승수법을 이용해 문제를 풀 수 있다.
- 위와 같은 식으로 $f(w)$와 $h(w) = 0 $를 하나의 식을 정의할 수 있다.

- 여기서 $\beta$를 lagrange multipliers라고 부름
- 우린 편미분을 통해 변화량이 0인 지점인 w와 $\beta$를 찾아낼 것이다.

`-` generalized Lagrangian (inequality)

- 조건이 =0이 되는 것뿐만 아닌 부등식일 경우에도 제한된 최적화문제로 일반화 할 것이다.
- 위의 식들은 primal optimization problem이라고 함

- 마찬가지로 라그랑지안을 사용하여 풀어낼 수 있음

`-` 그런데 왜 Lagrangian을 이용하나요?


- 제약 조건이 있는 다변수 함수의 최적화 문제를 제약 조건이 없는 형태로 바꾸어 푸는 방법
- 출처https://decisionboundary.tistory.com/2

- 따라서 우리는 f(w)를 최소화하고 싶었고 이것은  $\alpha$, $\beta$를 최대화해서 빼면 이것은 우리가 찾는 목적함수일 것이다.
- 추가적으로 졔약조건에 부합하지 않는다면 우리가 적용할 수 목적함수는 무한해진다,.

- 반대로 조건에 만족이 된다면 우리가 찾게 되는 것은 결국!!
- 조건이 만족된 $\theta_p(w)$는 결국 우리가 본래 알고싶었던 $f(w)$를 찾는 것과 같다

- $\theta_p(w)$ = $max\mathcal{L}(w,\alpha,\beta)$ 이였으므로 이들 minimize한다는 것은 
- 위의 식처럼 정의할 수 있다.

---

- $p^*$= $min_w\theta_p(w)$
- 이제 다른문제를 한번 봐보자
- 바로 $\theta_D(\alpha,\beta)= min_w\mathcal{L}(w,\alpha,\beta)$

- $ \mathcal{D}$ 는 dual이란 뜻
- $\alpha, \beta$를 최대화(최적화)한다는 것은 w를 최소화한다는 의미이다.
- 따라서 이러한 표기가 가능해짐

- 맨 앞서 보여줬던 $p^*$와 최대화, 최소화가 정확히 반대로 되는 것이다
- 또한 우리는 $d^*$를 $max_{\alpha, \beta : \alpha \ge 0 }\theta_p(w)$로 두고
- $p^*$와 $d^*$의 관계를 보자

- max min <= min max의 직관적이해


- 결국 우리는 최적화 문제를 듀얼 문제로 해결할 수 있다.
- 그렇지만 조건들이 있음
    - convex한 문제여야 함 그렇지 않으면 local minimum이 너무 많음
    - h는 affine(?)
    - 또한 g는 엄격하게 실현 가능하다고 가정할 것, 이는 $g_i(w)$가 모든 i에 대해 0보다 작다는 것을 의미

---

`-` KKT





- 위의 가정하에서 $w^*$에 관해서 $w^*,\alpha^*,\beta^*$가 존재해야하는 것이 가장 우선적인 문제이고
-  $\alpha^*,\beta^*$가 듀얼문제의 해결책이며
- 그렇다면 더 나아가 $p^* = d^* = \mathcal{L}(w^*,\alpha^*,\beta^*)$로 유도할 수 있다
- 또한 우리는 위의 3가지의 모수를 KKT 조건에 만족한다고 볼 것임

# 2. Optimal margin classifiers: the dual form

- 앞서 최적화를 위해 이러한 공식을 제시했고
- 제약 조건을 위처럼 다시 쓸 수 있다.

- KKT dual complementarity condition에서 functional margin이 1과 같은 훈련 셋에서만 $\alpha  > 0$ 가질 것이다.

```python
- 여기서 decision boundary와 가장 가까운 3개의 점이 support vectors라고 불림
- 많은 훈련데이터중 극히 일부만 쓰임(이상치에 둔감하다)
```

- 우리는 듀얼 폼의 문제를 개발할 때, 주의해야할 핵심 아이디어는 입력 피쳐 공간의 점 사이에 내부 곱으로 알고리즘을 표현할 수 있다는 것이었다.
- 해당문제는 부등식 제약만 있기 때문에 $\beta_i$는 없음

- 우선 w에 대한 $\mathcal{L}$을 최소화하고 $\theta_{\mathcal{D}}$를 구함
- 이는 w, b를 0으로 두는 $\mathcal{L}$의 편미분을 구하면 된다

따라서 앞서 구한 w가 정의된 식에 위에 것들을 대입하여 공식을 풀면 이렇게 정리가 되고 
마지막 자리는 0이되어 사라지게된다

- w, b를 반영한 목적함수를 최소함으로써 앞서 보여줬던 방정식을 도출했음
- 또한 제약조건들을 전부 합하면
- 우리는 듀얼 최적화 문제를 얻을 수 있음

- 우리는 앞서 제약조건에 따라 w(a)를 최대화하는 a를 찾는 것으로 최적의 w를 찾을 수 있음
- 따라서 절편항 b를 찾는 것은 간단해진다

- 다음으로 넘어가기전에 $\alpha$사이의 최적화하기 위한 w에 대한 방정식은 
- 모형의 훈련데이터에 맞추고 새로운 데이터를 예측하기 원할 때 
- $w^T+b$를 계산하고 해당 크기가 0보다 크다면 y를 1로 예측할 것이다
- 이것은 위처럼 나타낼수 있음

- 우리가 $\alpha$를 찾았다면 예측을 위해, x와 훈련데이터 사이에 내적에 의존하는 크기를 계산해야한다
- 또한 서포트벡터를 제외한 $\alpha$들은 전부 0이 될것임 따라서 많은 항들이 0이 될 것이고 적은 수의 서포트 벡터만을 가지고 예측이 가능하다!
- 최적화 문제의 이중형태를 알아냄으로써 입력 특징 벡터들 사이의 내적 측면에서만 전체 알고리즘을 작성할 수 있었다
- 이것은 더 높은 차원의 공간에서 효율적으로 학습할 수 있다!

# 4. SMO

- SVM의 듀얼 문제를 효율적으로 해결할 수 있는 방법을 제공

`-` coordinate ascent

- 이 알고리즘의 가장 안쪽 루프에서 우리는 모든 변수를 유지할 것
- 일부 $\alpha$를 제외하고 매개변수 $\alpha$에 대한 W를 다시 최적화
- 어떤 것이 w(a)를 가장 많이 증가시킬 것으로 예상하는지에 따라 업데이트할
다음변수를 채택

- 그렇다면 a2,,,an을 고정하고 coordinate ascent step을 취한다음 다시 a1에 대한 목표를 최적화한다면?
- 우리는 어떠한 진전을 이루기 어려움
- 그 이유가 위의 식과 같다


- $\alpha$들은 서로가 영향을 미치며 결정됨 
- 만약 다른 $\alpha$를 고정시킨다면 $\alpha_1$를 변화시킬 수가 없다!
- 따라서 업데이트를 원한다면 적어도 제약조건을 동시에 만족하는 2개의 $\alpha$를 업데이트해야함

- 그래서 우리는 이러한 룰을 따를 것이다.
- 우리는 글로벌 맥시범을 향한 가장 큰 진적을 이루는 $\alpha$ pair를  선택할 것
- 모든 값을 유지하고 $\alpha$ pair에 대해 다시 W(a)를 최적화

- 이 알고리즘의 수렴을 테스트하기 위해 
- 우리는 tol 사이의 KKT 조건에 만족하는지를 확인할 수 있다.
- tol은 the convergence tolerance parameter이다.(이 값보다 작아질때까지 최적화)

- 우리는 조건을 만족하며 $\alpha$들을 이렇게 셋팅할 수 있다.
- $\alpha_3,,,\alpha_n$까진 고정되며 $\alpha_1,\alpha_2$를 반영하여 새롭게 최적화 할 것임

- 우린 제약조건에 따라 $\alpha$는 C범위 안에 있어야하는 것을 알고 있음
- 또한 $\alpha_1,\alpha_2$의 선형적인 형태는 C로 이루어진 박스를 관통하고 있다.
- 그렇다면 이를 허용할 수 있는 가장 작은 L과 최대 H가 있을 것
