# 15.1  Markov decision processes

- S : states의 집합
- A : actions의 집합
- P : the state transition porbabilities
- $\gamma$ : discount factor
- R : S $\times$ A $\rightarrow \mathbb{R}$ reword function

- MDP 과정의 변화
- state $s_0$에서 시작 
- 어떠한 action $a_0$을 선택함 
- $s_1$으로 state가 변경되어지고 이는 $s_1 \sim P_{s_0a_0}$ 
- 그리고 다시 어떠한 action $a_1$이 선택되어짐
- 위 과정이 반복

- 또한 행동의 결과에 따른 보상은 위와 같이 주어진다.
- reword function에 $\gamma$ discount factor가 곱해진 형태

`-` what we have the goal?

- 강화학습의 목표는 the total payoff의 기대값이 최대화되는 것을 원함

- 시간에 따른 보상은 $\gamma^t$만큼 할인됨
- 따라서 보상을 크게 얻고 싶으면 긍정적인 보상을 가능한 빨리 얻으려고 할거고
- 반대로 부정적인 보상은 최대한 미루게 될것임  

- Policy라
    - $S \rightarrow A$로 mapping해줌 (the states로부터 actions으로)
    - 어떠한 Policy가 실행된다는 건 state s에 있을 때마다 
    - action $a = \pi(s)$을 취한다는 의미
    - 이것을 policy $\pi$에 따른 value function이라고 정의할것임

- $V^\pi(s)$는 state s에서 $\pi$를 따르는 액션을 취하면서 시작되는 할인된 보상의 기대 합
- policy $\pi$가 주어진다면 value function $V^\pi$는 Bellman equations을 만족

- 벨만 방정식은 현재 상태의 가치와 다음 상태의 가치 사이의 관계를 보여줌
- 고정된 정책이 있다면 그 정책을 따라 항상 같은 행동을 할 것임

- 이것들은 $V^\pi(s)$ 할인 보상의 예상 합은 두 개의 항으로 이루어져 있음
- first, state s가 시작되는 동안의 바로 얻을 수 있는 즉각적인 보상
- second, 미래의 할인된 보상의 예상된 합임.
- 첫번째 단계 이후에 얻은 할인된 보상의 기대합



- 밸만 방정식은 value function을 효과적으로 풀 수있음
- 특히 유한한 공간일 때, 모든 state s에 대한 $V^\pi(s)$를 전부 쓸 수가 있다.
- 이것들은 선형적인 식을 줄 것임

- 어떠한 policy를 이용해 가능한 기대되는 보상의 최고 합
- 취할수 있는 액션중에서 가장 큰 state value function을 취하는 쪽으로 선택
- optimal value function

- $\pi^*(s)$ : 매순간 state에서 action을 취할 때 가장 최고의 action을 취한다는 의미

- $\pi(s)$는 max에서 최대값에 달하는 a를 줄것임
- 이것은 모든 state와 policy에서 위와같은 식을 가질 수 있는 사실이된다
- 최적의 $V^*(s)$는  $V^\pi(s)$와 같고 이들은 어떤 다른 정책보단 값이 클것

- state가 어떻든 최적의 policy가 있을 것
- 따라서 같은 policy $\pi$는 모든 상태 s에 대해 15.1식에 대해 최대값을 얻음

# 15.2 Value iteration and policy iteration

- 유한한 MDPs문제를 해결하는 효율적 알고리즘 2개

- 이 알고리즘은 벨만 방정식을 이용해 추정 함수를 반복적으로 업데이트 하는 것
- 업데이트를 하는 방법은 2가지이다
    - 모든 상태에 대한 $V(s)$에 관한 새 값들을 계산한 다음 모든 이전 값을 새값으로 덮어씀
    (synchronous update), 이 경우 알고리즘은 현재 추정치를 사용하는 벨만 백업 연산자를 구현하는 것으로 볼 수 있다
    
    - 또는 비동기 업데이트(asynchronous update)가 있다. 한번에 하나씩 값을 업데이트 하면서 상태를 루프오버함
- 이들로 인해 V가 V*으로 수렴함

- value iteration말고 optimal policy를 찾는 두번째 표준 알고리즘이 있다.
- policy iteration

    - b단계 : greedy with respect to V

# 15.3   Learning a model for an MDP

- 지금까지는 상태 전이 확률과 보상이 알려져 있다고 가정한 MDPs를 논의 했음
- 많은 현실적인 문제에서는 이 확률과 보상이 명시적으로 나타나지 않는다
- 그러나 데이터를 통해 추론해야함 (usually, S, A and $\gamma$ are known)

`-` example : inverted pendulum problem

https://www.youtube.com/watch?v=nOSTzpA0nGk

- $s_i^{(j)}$는 j의 시도의 i점 시점에 우리의 state이다. 
- $a_i^{(j)}$는 해당 state로 부터 취해진 대응 액션이다.
- 각각의 시도는 MDP가 종료할 때까지 실행되며 크고 유한한 타임스텝으로 실행됨

- 다수의 시도로 통한 MDP의 경험은 우리에게 쉽게 최대 가능성 추정치를 줄 수 있음
- $P_{sa}(s')$
    - 분자 : state s에서 action a를 취할 때 s'를 얻는 횟수
    - 분모 : state s에서 action a를 취하는 횟수

- 만약 위의 비율이 0/0이라면(이전 상태 a에서 조치를 취한 적이 없는 경우)
- 우리는 단순히 $P_{sa}(s')$를 $1/|S|$로 추정할 수 있음($P_{sa}(s')$를 모든 상태에서 균일한 분포로 추정)

- 만약 우리가 더 많은 경험을 얻으면, 새로운 경험을 사용해 추정된 상태전이확률을 갱신하는 효율적 방법이 있다
- 구체적으로 분모 분자의 카운트를 유지한다면 계속 카운트가 축적이 될 수 있음
- 이 비율을 계산하는 것은 추정된 상태전이 확률이 주어지게된다.

1. 임의로 $\pi$를 초기화
2. 반복
    - 몇번의 시도를 통해 MDP내에서 $\pi$를 실행
    - MDP에서 축적된 경험을 사용하여 $P_{sa}$에 대한 추정치를 업데이트
    - 추정된 상태전환 확률 및 보상을 사용하여 값 반복을 적용하여 새로운 추정 값 함수를 얻음
    - $\pi$를 V를 반영한 greedy policy로 업데이트

- V = 0으로 초기화하는 것 대신에 
- 이전에 반복동안 해결된 방법으로 초기화한다면 
- 그 이후에 더 나은 초기 시작점으로 value iteration을 제공하고 더 빨리 수렴하게 될것임

```python
import numpy as np

# 상태 공간, 행동 공간, 전이 확률, 보상 설정
states = ['A', 'B', 'C', 'D', 'E']
actions = ['Left', 'Right']
transitions = {
    'A': {'Left': 'A', 'Right': 'B'},
    'B': {'Left': 'A', 'Right': 'C'},
    'C': {'Left': 'B', 'Right': 'D'},
    'D': {'Left': 'C', 'Right': 'E'},
    'E': {'Left': 'D', 'Right': 'E'}
}
rewards = {
    'A': {'Left': -3, 'Right': -1},
    'B': {'Left': -1, 'Right': 3},
    'C': {'Left': -1, 'Right': -1},
    'D': {'Left': 3, 'Right': -1},
    'E': {'Left': -1, 'Right': -3}
}

# 초기 정책 설정 (랜덤)
policy = {}
for state in states:
    policy[state] = np.random.choice(actions)

# 정책 평가 (가치 함수 계산)
def evaluate_policy(policy, transitions, rewards, gamma=0.9, theta=0.01):
    V = {state: 0 for state in states}
    
    while True:
        delta = 0
        for state in states:
            v = V[state]
            action = policy[state]
            next_state = transitions[state][action]
            reward = rewards[state][action]
            V[state] = reward + gamma * V[next_state]
            delta = max(delta, abs(v - V[state]))
        
        if delta < theta:
            break
    
    return V

# 정책 개선
def improve_policy(policy, transitions, rewards, V, gamma=0.9):
    stable = True
    for state in states:
        old_action = policy[state]
        max_action = None
        max_value = float('-inf')
        for action in actions:
            next_state = transitions[state][action]
            value = rewards[state][action] + gamma * V[next_state]
            if value > max_value:
                max_value = value
                max_action = action
        policy[state] = max_action
        if old_action != max_action:
            stable = False
    
    return stable

# Policy Iteration
while True:
    V = evaluate_policy(policy, transitions, rewards)
    if improve_policy(policy, transitions, rewards, V):
        break

# 결과 출력
print("Optimal Policy:")
for state in states:
    print(f"State {state}: Choose action {policy[state]}")
```
